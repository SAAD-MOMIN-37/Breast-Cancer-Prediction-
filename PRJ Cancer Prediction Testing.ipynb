**Loading Models** 

```python
with open('models/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
with open('models/label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)
with open('models/feature_names.pkl', 'rb') as f:
    feature_names = pickle.load(f)
with open('models/voting_classifier.pkl', 'rb') as f:
    voting_model = pickle.load(f)

print("✓ Models loaded successfully")
```

**Prediction Function** 

```python
def predict_cancer(features_dict):
    features_df = pd.DataFrame([features_dict], columns=feature_names)
    features_scaled = scaler.transform(features_df)
    
    prediction = voting_model.predict(features_scaled)[0]
    probability = voting_model.predict_proba(features_scaled)[0]
    
    diagnosis = label_encoder.inverse_transform([prediction])[0]
    
    return {
        'diagnosis': 'Malignant' if diagnosis == 'M' else 'Benign',
        'confidence': float(probability[prediction]),
        'prob_benign': float(probability[0]),
        'prob_malignant': float(probability[1])
    }
```

**Test Prediction** 

```python
def simple_cancer_prediction():
    """Simplified cancer prediction with minimal user input"""
    
    print("\n" + "="*70)
    print("BREAST CANCER PREDICTION SYSTEM")
    print("="*70)
    print("\nType 'exit' to quit\n")
    
    while True:
        try:
            print("\n" + "-"*70)
            user_input = input("Press Enter to continue or type 'exit' to quit: ").strip()
            
            if user_input.lower() in ['exit', 'quit', 'q']:
                print("\nGoodbye!\n")
                break
            
            print("\nEnter your tumor measurements:\n")
            
            # Get essential features
            features = {}
            
            print("Enter the following values:")
            features['radius_mean'] = float(input("  Radius Mean (6-30): "))
            features['texture_mean'] = float(input("  Texture Mean (9-40): "))
            features['perimeter_mean'] = float(input("  Perimeter Mean (40-200): "))
            features['area_mean'] = float(input("  Area Mean (140-2500): "))
            features['smoothness_mean'] = float(input("  Smoothness Mean (0.05-0.16): "))
            features['compactness_mean'] = float(input("  Compactness Mean (0.01-0.35): "))
            features['concavity_mean'] = float(input("  Concavity Mean (0-0.43): "))
            features['concave points_mean'] = float(input("  Concave Points Mean (0-0.20): "))
            features['symmetry_mean'] = float(input("  Symmetry Mean (0.10-0.30): "))
            features['fractal_dimension_mean'] = float(input("  Fractal Dimension Mean (0.04-0.10): "))
            
            # Auto-fill other features
            se_defaults = {
                'radius_se': 0.405, 'texture_se': 1.217, 'perimeter_se': 2.866,
                'area_se': 40.34, 'smoothness_se': 0.007, 'compactness_se': 0.026,
                'concavity_se': 0.032, 'concave points_se': 0.012,
                'symmetry_se': 0.020, 'fractal_dimension_se': 0.004
            }
            
            for key in list(features.keys()):
                worst_key = key.replace('_mean', '_worst')
                features[worst_key] = features[key] * 1.4
            
            features.update(se_defaults)
            
            # Prepare features for prediction
            features_df = pd.DataFrame([features], columns=feature_names)
            features_scaled = scaler.transform(features_df)
            
            # Predict using voting model
            prediction = voting_model.predict(features_scaled)[0]
            probability = voting_model.predict_proba(features_scaled)[0]
            
            diagnosis = label_encoder.inverse_transform([prediction])[0]
            diagnosis_text = 'Malignant' if diagnosis == 'M' else 'Benign'
            
            print("\n" + "="*70)
            print("PREDICTION RESULT")
            print("="*70)
            
            if diagnosis_text == 'Malignant':
                print(f"\n[MALIGNANT] Diagnosis: {diagnosis_text}")
            else:
                print(f"\n[BENIGN] Diagnosis: {diagnosis_text}")
            
            print(f"Confidence: {probability[prediction]*100:.2f}%")
            print(f"Probability Benign: {probability[0]*100:.2f}%")
            print(f"Probability Malignant: {probability[1]*100:.2f}%")
            
            print("\nDISCLAIMER: For educational purposes only.")
            print("Consult a medical professional for actual diagnosis.")
            print("="*70)
            
        except ValueError:
            print("\nERROR: Invalid input! Please enter numeric values.")
        except KeyboardInterrupt:
            print("\n\nGoodbye!\n")
            break
        except Exception as e:
            print(f"\nERROR: {e}")

# Run simple prediction
simple_cancer_prediction()
```
